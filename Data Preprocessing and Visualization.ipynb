```python
# Perform Chi-Square and Monte Carlo testing
# Null hypothesis (H0): There is no significant relationship between customer state and popular product category.
# Convert the purchase history data into a list and store it in the 'data' variable
data = list(purchase_history_collection.find())

# Transform the data into a pandas DataFrame
df = pd.DataFrame(data)

# Verify that the 'product_category_name' and 'customer_state' columns exist in the DataFrame
if 'product_category_name' in df.columns and 'customer_state' in df.columns:

    # Create a contingency table for 'product_category_name' and 'customer_state'
    contingency_table = pd.crosstab(df['product_category_name'], df['customer_state'])

    # Perform the Chi-Square Test
    chi2_obs, p_obs, dof, expected = chi2_contingency(contingency_table)

    # Calculate CramÃ©r's V to measure effect size
    n = contingency_table.sum().sum()  
    cramers_v = sqrt(chi2_obs / (n * (min(contingency_table.shape) - 1)))

    # Display Chi-Square results, p-value, and CramÃ©r's V
    print("Observed Chi-Square Statistic:", chi2_obs)
    print("Observed p-value:", p_obs)
    print("CramÃ©râ€™s V (effect size): {:.5f}".format(cramers_v))
    print("Degrees of Freedom:", dof)
    print("Expected Frequencies:\n", expected)

    # Perform Monte Carlo simulation with 500,000 iterations
    num_simulations = 500000  
    chi2_simulated = []

    # Simulate Chi-Square statistics based on random contingency tables
    for _ in range(num_simulations):
        simulated_table = np.random.multinomial(
            contingency_table.sum().sum(), 
            np.ones(contingency_table.size) / contingency_table.size  
        ).reshape(contingency_table.shape)

        # Calculate the Chi-Square statistic for the simulated table
        chi2_sim, _, _, _ = chi2_contingency(simulated_table)
        chi2_simulated.append(chi2_sim)

    # Convert the list of simulated Chi-Square statistics to a NumPy array
    chi2_simulated = np.array(chi2_simulated)

    # Calculate the Monte Carlo p-value
    p_monte_carlo = np.mean(chi2_simulated >= chi2_obs)

    # Display Monte Carlo simulation results
    print("Monte Carlo p-value: {:.5f}".format(p_monte_carlo))

    # Plot the distribution of simulated Chi-Square statistics
    plt.hist(chi2_simulated, bins=50, color='lightblue', edgecolor='black', alpha=0.7)
    plt.axvline(chi2_obs, color='red', linestyle='dashed', linewidth=2, label=f'Observed chi2 = {chi2_obs:.2f}')
    plt.title('Distribution of Simulated Chi-Square Statistics')
    plt.xlabel('Chi-Square Statistic')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    # Interpret results based on a significance level of 0.05
    alpha = 0.05  

    if p_monte_carlo < alpha:
        print("Reject the null hypothesis (H0): Customer state and product category have a significant relationship.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant relationship between customer state and product category.")
else:
    print("Error: 'product_category_name' and 'customer_state' columns are missing from the DataFrame.")
```

```python
def display_category_frequency(year, purchase_history_collection):
    pipeline = [
        {
            "$group": {
                "_id": "$product_category_name",
                "count": {"$sum": 1}
            }
        },
        {
            "$sort": {"count": -1}
        }
    ]

    # Execute aggregation pipeline
    category_frequency = list(purchase_history_collection.aggregate(pipeline))

    # Calculate the total number of purchases
    total_count = sum(category['count'] for category in category_frequency)

    # Prepare to display the data in table format
    table_data = []
    for index, category in enumerate(category_frequency, start=1):
        percentage = (category['count'] / total_count) * 100
        table_data.append([index, category['_id'], category['count'], f"{percentage:.2f}%"])

    # Define table headers, including the 'Ranking' column
    headers = ["Ranking", "Category", "Frequency", "Percentage"]

    # Display the table with a simple structure (no borders)
    print(f"\n{year} Purchase history popularity ranking")
    print(tabulate(table_data, headers, tablefmt="simple"))

display_category_frequency("2016 to 2018", purchase_history_collection)
display_category_frequency("2016", year_2016_purchase_history_collection)
display_category_frequency("2017", year_2017_purchase_history_collection)
display_category_frequency("2018", year_2018_purchase_history_collection)
```

```python
# Function to calculate and display the mean product category frequency for a specified year or range of years
def calculate_mean_product_frequency(year, purchase_history_collection):
    pipeline = [
        {
            "$group": {
                "_id": "$product_category_name",
                "count": {"$sum": 1}
            }
        },
        {
            "$sort": {"count": -1}  
        }
    ]

    category_frequency = list(purchase_history_collection.aggregate(pipeline))

    total_count = sum(category['count'] for category in category_frequency)
    num_categories = len(category_frequency)

    mean_frequency = total_count / num_categories if num_categories > 0 else 0
    mean_percentage = (mean_frequency / total_count) * 100 if total_count > 0 else 0

    print(f"{year} purchase history")
    print(f"Average product purchase(number): {mean_frequency:.2f}")
    print(f"Average product purchase (percentage): {mean_percentage:.2f}%")
    print("\n")

calculate_mean_product_frequency("2016 to 2018", purchase_history_collection)
calculate_mean_product_frequency("2016", year_2016_purchase_history_collection)
calculate_mean_product_frequency("2017", year_2017_purchase_history_collection)
calculate_mean_product_frequency("2018", year_2018_purchase_history_collection)
```

```python
def calculate_median_product_frequency(year_range, purchase_history_collection, year=None):
    # Optional: Add a filter for specific years
    match_stage = {}
    if year:
        match_stage = {
            "$match": {
                "$expr": {"$eq": [{"$year": {
                    "$dateFromString": {"dateString": "$order_purchase_timestamp"}
                }}, year]},
                "product_category_name": {"$ne": None, "$ne": ""}
            }
        }
    
    # Pipeline to group and count product categories
    pipeline = []
    if match_stage:
        pipeline.append(match_stage)
    
    pipeline.extend([
        {
            "$project": {
                "product_category_name": {"$toLower": "$product_category_name"}
            }
        },
        {
            "$group": {
                "_id": "$product_category_name",
                "count": {"$sum": 1}
            }
        },
        {
            "$sort": {"count": -1}
        }
    ])

    category_frequency = list(purchase_history_collection.aggregate(pipeline))
    
    if not category_frequency:
        print(f"No data found for {year_range}.")
        return
    
    total_count = sum(category['count'] for category in category_frequency)
    frequencies = [category['count'] for category in category_frequency]

    # Calculate median frequency
    frequencies.sort()  
    n = len(frequencies)

    if n % 2 == 1:
        median_frequency = frequencies[n // 2]  # Middle value for odd
    else:
        median_frequency = (frequencies[n // 2 - 1] + frequencies[n // 2]) / 2  # Average for even

    # Check for all categories that match the median frequency
    median_categories = [category for category in category_frequency if category['count'] == median_frequency]
    
    print(f"\n{year_range} purchase history")
    
    if median_categories:
        # Collect names of all categories with the median frequency
        median_category_names = [category['_id'] for category in median_categories]
        # Join the names with a comma
        median_category_display = ', '.join(median_category_names)

        # Calculate percentage for any one of the found categories (using the first as representative)
        median_percentage = (median_categories[0]['count'] / total_count) * 100 if total_count > 0 else 0

        print(f"Categories with Median Frequency: {median_category_display}")
        print(f"Frequency: {median_categories[0]['count']}")
        print(f"Percentage: {median_percentage:.2f}%")
    else:
        # If no exact match is found, find the closest category to the median frequency
        closest_category = min(category_frequency, key=lambda x: abs(x['count'] - median_frequency))
        closest_percentage = (closest_category['count'] / total_count) * 100 if total_count > 0 else 0
        
        print(f"Closest Category: {closest_category['_id']}")
        print(f"Frequency: {closest_category['count']}")
        print(f"Percentage: {closest_percentage:.2f}%\n")

# Example usage
calculate_median_product_frequency("2016 to 2018", purchase_history_collection)
calculate_median_product_frequency("2016", year_2016_purchase_history_collection)
calculate_median_product_frequency("2017", year_2017_purchase_history_collection)
calculate_median_product_frequency("2018", year_2018_purchase_history_collection)
```

```python
from pymongo import MongoClient
from datetime import datetime, timedelta

pipeline_weekly = [
    {
        # Step 1: Convert `order_purchase_timestamp` string to date format
        "$addFields": {
            "purchase_date": {
                "$dateFromString": {
                    "dateString": "$order_purchase_timestamp",
                    "format": "%Y-%m-%d %H:%M:%S"  # Matching the given format
                }
            }
        }
    },
    {
        # Step 2: Group by week of the purchase date
        "$group": {
            "_id": {
                "category": "$product_category_name",  # Group by product category
                "year": {"$year": "$purchase_date"},  # Extract the year
                "week": {"$isoWeek": "$purchase_date"}  # Extract ISO week number
            },
            "purchase_count": {"$sum": 1}  # Count of purchases
        }
    },
    {
        # Step 3: Sort by year and week
        "$sort": {"_id.year": 1, "_id.week": 1}
    }
]

# Execute the aggregation pipeline
weekly_trends = list(year_2016_purchase_history_collection.aggregate(pipeline_weekly))

# Check if there are any results from the aggregation
if not weekly_trends:
    print("No trends found for the specified collection.")
else:
    # To store the trends by week
    trends_by_week = {}

    # Populate the trends_by_week dictionary
    for trend in weekly_trends:
        year = trend['_id']['year']
        week = trend['_id']['week']
        category = trend['_id']['category']
        purchase_count = trend['purchase_count']

        # Create a key for the week
        week_key = f"{year} (Week {week})"

        # Initialize if the week is not in the dictionary
        if week_key not in trends_by_week:
            trends_by_week[week_key] = []

        # Append the category and purchase count to the respective week
        trends_by_week[week_key].append((category, purchase_count))

    # Find the first and last weeks present in the dataset
    week_keys = list(trends_by_week.keys())
    
    # Ensure week_keys is not empty before proceeding
    if week_keys:
        first_week_key = min(week_keys, key=lambda x: (int(x.split(' (Week ')[0]), int(x.split(' (Week ')[1][:-1])))  # Corrected line
        last_week_key = max(week_keys, key=lambda x: (int(x.split(' (Week ')[0]), int(x.split(' (Week ')[1][:-1])))  # Corrected line
        # Get the first week number and year from the first week key
        first_year, first_week = map(int, first_week_key.replace(')', '').split(' (Week '))
        last_year, last_week = map(int, last_week_key.replace(')', '').split(' (Week '))

        # Generate weeks in the range from the first to the last week
        def generate_weeks(first_week_key, last_week_key):
            """Yield week keys between first and last week."""
            # Remove the closing parenthesis and split properly
            first_year, first_week = map(int, first_week_key.replace(')', '').split(' (Week '))
            last_year, last_week = map(int, last_week_key.replace(')', '').split(' (Week '))

            current_date = datetime.strptime(f"{first_year}-W{first_week}-1", "%Y-W%W-%w")

            # Loop until we reach the last week
            while current_date <= datetime.strptime(f"{last_year}-W{last_week}-1", "%Y-W%W-%w"):
                yield f"{current_date.year} (Week {current_date.isocalendar()[1]})"
                current_date += timedelta(weeks=1)

        # Check if there is any data found
        weeks_data_found = False  # To track if any week has data
        no_data_start_week = 1  # The first week of the year

        # If the first week is greater than 1, we need to report that there's no data from Week 1 to the first available week
        if first_week > no_data_start_week:
            print(f"There is no data from {first_year} (Week {no_data_start_week}) to {first_year} (Week {first_week-1}).")

        # Initialize variables to track ranges of weeks with no data
        no_data_start = None

        # Display the results in the desired format
        for week_key in generate_weeks(first_week_key, last_week_key):
            if week_key in trends_by_week:
                if no_data_start is not None:
                    # If we were tracking a no data range, print it
                    if no_data_start == week_key:  # Single week case
                        print(f"\nThere is no data for {no_data_start}.")
                    else:  # Range case
                        last_week_no = int(week_key.strip().split(' ')[-1].replace('(', '').replace('Week ', '').replace(')', '')) - 1
                        last_week_key = f"2016 (Week {last_week_no})"
                        print(f"\nThere is no data from {no_data_start} to {last_week_key}.")

                    no_data_start = None  # Reset

                categories = trends_by_week[week_key]

                # Find the maximum purchase count for the week
                max_count = max(count for _, count in categories)

                # Filter categories with the maximum purchase count
                trending_categories = [category for category, count in categories if count == max_count]

                # Print results
                print(f"\n{week_key} \nTrend Category:")
                for category in trending_categories:
                    print(f"{category} ({max_count} purchases)")

                weeks_data_found = True
            else:
                # We found a week with no data, start tracking the range
                if no_data_start is None:
                    no_data_start = week_key  # Start of no data range

        # Check if we ended with an unclosed no data range
        if no_data_start is not None:
            if no_data_start == week_key:  # Single week case
                print(f"There is no data for {no_data_start}.")
            else:  # Range case
                print(f"There is no data from {no_data_start} to {week_key}.")

        print()  # Print a newline for better separation between weeks

        # Explicitly check for week 52
        if "2016 (Week 52)" not in trends_by_week:
            print("There is no data for 2016 (Week 52).")

        # If no data was found for the entire year
        if not weeks_data_found:
            print("There is no data for the entire year.")

    else:
        print("No week keys available to process.")
```
