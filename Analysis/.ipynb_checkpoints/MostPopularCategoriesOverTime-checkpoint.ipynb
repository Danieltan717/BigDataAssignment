```python
import pandas as pd

# Load the cleaned dataset
full_data = pd.read_csv('cleaned_full_data.csv')

# Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

# Create a new column for the season
full_data['season'] = full_data['order_month'].apply(get_season)

# Group by season and category to count orders
seasonal_popularity = full_data.groupby(['season', 'product_category_name_english'])['order_id'].count().reset_index()

# Identify the highest category for each season
highest_category_per_season = seasonal_popularity.loc[seasonal_popularity.groupby('season')['order_id'].idxmax()]

# Display the result
print(highest_category_per_season[['season', 'product_category_name_english', 'order_id']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
full_data = pd.read_csv('cleaned_full_data.csv')

# Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

# Create a new column for the season
full_data['season'] = full_data['order_month'].apply(get_season)

# Group by season and category to count orders
seasonal_popularity = full_data.groupby(['season', 'product_category_name_english'])['order_id'].count().reset_index()

# Sort values for better visualization
seasonal_popularity = seasonal_popularity.sort_values(by=['season', 'order_id'], ascending=[True, False])

# Display the top categories for each season
top_categories_per_season = seasonal_popularity.groupby('season').head(5)

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(data=top_categories_per_season, x='order_id', y='product_category_name_english', hue='season', dodge=True, palette='viridis')
plt.title('Top 5 Product Categories by Season')
plt.xlabel('Number of Orders')
plt.ylabel('Product Category')
plt.legend(title='Season')
plt.grid(axis='x')
plt.show()
```
