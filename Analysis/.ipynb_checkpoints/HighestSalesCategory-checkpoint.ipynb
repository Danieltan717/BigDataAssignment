```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
full_data = pd.read_csv('cleaned_full_data.csv')

# Convert price to numeric
full_data['price'] = pd.to_numeric(full_data['price'], errors='coerce')

# Calculate total sales for each category
category_sales = full_data.groupby('product_category_name_english')['price'].sum().reset_index()

# Sort categories by sales for better visualization
category_sales = category_sales.sort_values(by='price', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 26))
sns.barplot(x='price', y='product_category_name_english', data=category_sales, hue='product_category_name_english', palette='viridis', legend=False)
plt.title('Total Sales by Product Category')
plt.xlabel('Total Sales (in currency)')
plt.ylabel('Product Category')
plt.grid(axis='x')
plt.show()
```
