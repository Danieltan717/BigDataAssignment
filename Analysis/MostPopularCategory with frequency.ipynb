```python
# Import necessary libraries
import pandas as pd

# Load the cleaned dataset
full_data = pd.read_csv('cleaned_full_data.csv')

# Step 1: Count the frequency of each product category
category_frequency = full_data['product_category_name_english'].value_counts()

# Step 2: Convert the result to a DataFrame for better readability
category_frequency_df = category_frequency.reset_index()
category_frequency_df.columns = ['product_category_name_english', 'frequency']

# Step 3: Select the top 5 most popular categories
top5_categories = category_frequency_df.head(5)

# Step 4: Display the top 5 most popular categories
print(top5_categories)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 20))
plt.barh(category_frequency_df['product_category_name_english'], category_frequency_df['frequency'], color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Product Category')
plt.title('Frequency of Product Categories')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest frequency at the top
plt.tight_layout()  # Adjust layout for better spacing
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Group by month and category
monthly_trends = full_data.groupby(['order_year', 'order_month', 'product_category_name_english'])['order_id'].count().reset_index()

# Create a date column from year and month
monthly_trends['date'] = pd.to_datetime(monthly_trends['order_year'].astype(str) + '-' + monthly_trends['order_month'].astype(str) + '-01')

# Plotting trends for a specific category
specific_category = 'bed_bath_table'  # Change this to any category
category_trend = monthly_trends[monthly_trends['product_category_name_english'] == specific_category]

plt.figure(figsize=(12, 6))
plt.plot(category_trend['date'], category_trend['order_id'], marker='o', linestyle='-', color='b')
plt.title(f'Trend of {specific_category} Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Purchases', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
